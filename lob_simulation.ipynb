# Reproducing stylized facts

```python
import wandb
from dotenv import load_dotenv
import os
import importlib
import wandb_sweep


sweep_config = {
    "method": "bayes",
    "metric": {
        "name": "err",
        "goal": "minimize"
    },
    "parameters": {
        "save_path": {"values": [None]},
        "N": {"values": [200, 500, 1000]},
        "w": {"min": 0.1, "max": 0.9},
        'Pf': {"values": [10000]},
        "S": {"values": [500, 1000, 1500]},
        "deltaP": {"values": [1]},
        "tc": {"values": [500, 1000, 1500, 2000]},
        "TE": {"values": [5000, 10000, 15000, 20000]},
        "L": {"values": [100, 200, 300]},
        "D": {"values": [5, 10, 20]},
        "R": {"values": [5]},

        "ALPHA": {"distribution": "log_uniform_values", "min": 0.01, "max": 0.9},
        "GAMMA": {"distribution": "uniform", "min": 0.7, "max": 0.99},
        "EPS0": {"distribution": "log_uniform_values", "min": 0.1, "max": 1.0},
        "EPS_DECAY": {"distribution": "log_uniform_values", "min": 0.90, "max": 0.999},
        "EPS_MIN": {"distribution": "log_uniform_values", "min": 0.01, "max": 0.1},

        "LAMBDA_INV": {"distribution": "log_uniform_values", "min": 0.001, "max": 0.9},
        "LAMBDA_HOLD": {"distribution": "log_uniform_values", "min": 0.001, "max": 0.9},
    }
}

load_dotenv()
secret_value = os.getenv("SECRET_VALUE")

wandb.login(key=secret_value)

importlib.reload(wandb_sweep)

sweep_id = wandb.sweep(
    sweep_config,
    project="lob-simulation-optimization"
)

wandb.agent(
    sweep_id,
    function=wandb_sweep.train,
    project="lob-simulation-optimization",
    count=20
)
```
